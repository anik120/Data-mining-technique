```python
def normal_distribution(mean, sigma, x):
        exponent = -float(((x-mean)**2))/(2*(sigma**2))
        return (1/(np.sqrt(2*3.14*(sigma**2)))) * np.exp(exponent)
```

```python
def gaussian(mean, sigma, N):
    accepted = 0 # the number of accepted samples
    samples = np.zeros(N)
    count = 0 # the total count of proposals

    while (accepted < N):
        x = np.random.normal(mean-sigma, mean+sigma)
        y = np.random.normal(0,1)/100
        if y < normal_distribution(mean, sigma, x):
            samples[accepted] = x
            accepted += 1
            
        count += 1
    print("Count", count, "Accepted", accepted)
    
    plt.hist(samples, bins = 300)
    plt.show()
```

```python
def uniform(min_, max_, n_samples):
    x = np.arange(min_, max_)
    mean = (min_ + max_)/2
    xU, xL = x + 0.5, x - 0.5 
    prob = ss.uniform.cdf(xU, loc = mean, scale = 100) - ss.uniform.cdf(xL, loc = mean, scale = 100)
    prob = prob / prob.sum() #normalize the probabilities so their sum is 1
    nums = np.random.choice(x, size = n_samples, p = prob)
    plt.hist(nums, bins = len(x))
    plt.show()
    return nums
```

```python
def normal_distribution_2d(mean, sigma, x):
        exponent = float(-((x[0]-mean[0])**2)/(2*sigma[0]**2)-((x[1]-mean[1])**2)/(2*sigma[1]**2))
        return (1/2*3.14*sigma[0]*sigma[1] ) * np.exp(exponent)
```

```python
def gaussian2d(mean, sigma, N):
    accepted = 0 # the number of accepted samples
    samples = []
    y_plot = []
    count = 0 # the total count of proposals

    while (accepted < N):
        x1 = np.random.normal(mean[0]-sigma[0], mean[0]+sigma[0])
        x2 = np.random.normal(mean[1]-sigma[1], mean[1]+sigma[1])
        x=[x1,x2]
        y = np.random.normal(0,1)/100
        if y < normal_distribution_2d(mean, sigma, x):
            samples.append(x[0])
            samples.append(x[1])
            y_plot.append(y)
            y_plot.append(y)
            accepted += 1
            
        count +=1
    print("Count", count, "Accepted", accepted)
    plt.scatter(samples, y_plot)
    plt.show()
```

```python
def genPopForStevens(m):
    pop = [];
    x = [];
    for i in range(m):
        temp = [i] * (m-i)*10;
        x = x + temp;
    pop = x;
    return pop;
```

```python
def getGroup(randNum,groupProb):
    group = 0;
    for group in range(len(groupProb)):
        if randNum < 20*groupProb[group]:
            break;
    return group
```

```python
def applyProbTest(meu,sigma):
    n = -10000;
    x = np.random.rand(2);
    z = np.sqrt(-2*np.log(x[0]))*np.cos(2*np.pi*x[1])
    n = np.float(z * sigma + meu)
    return n;

def getStevensSample(params):
    m = params[0];
    n = params[1];

    x = genPopForStevens(m);
#     print(x)
    xCounter = Counter(x);
#     print(xCounter) -----------
    #get probs
    popCount = np.sum(list(xCounter.values()));
    #print(popCount)
    probs = dict();
    for k in xCounter:
        probs[k] = float(xCounter[k])/popCount;
    #print(np.sum(np.array(list(probs.values()))));
    #get group probs
    print(probs)
    groupProb = []
    for eachGroup in range(int(m/n)):
        tempProb = 0;
        for i in range(n):
            tempProb += probs[eachGroup*n+i];
        tempProb /= n;
        for i in range(n):
            probs[eachGroup*n+i] = tempProb;
        groupProb.append(tempProb);
    #print(np.sum(np.array(list(probs.values()))));
    #print(groupProb)

    #get cumul. probs
    for eachProb in range(int(m/n)):
        if eachProb == 0:
            continue;
        #print(eachProb)
        groupProb[eachProb] += groupProb[eachProb - 1];
    #print(sum(np.array(groupProb)*20))
    #Assign frequencies to groups
    groupNum = [];
    for eachSample in range(n):
        randNum = np.random.rand();
        #print(randNum)
        groupNum.append(getGroup(randNum,groupProb));

    #print(groupNum)
    sample = [];
    for eachVal in groupNum:
        k = int(np.round(n*np.random.rand()+(eachVal*n)));
        while k in sample:
            k = int(np.round(n*np.random.rand()+(eachVal*n)));
        sample.append(k);
    return sample
```
