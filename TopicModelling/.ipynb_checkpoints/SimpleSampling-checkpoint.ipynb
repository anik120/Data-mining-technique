```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(mean, sigma, n_samples):
    x = np.arange(mean - sigma, mean + sigma)
    xU, xL = x + 0.5, x - 0.5 
    prob = ss.norm.cdf(xU, loc = mean, scale = sigma) - ss.norm.cdf(xL, loc = mean, scale = sigma)
    prob = prob / prob.sum() #normalize the probabilities so their sum is 1
    nums = np.random.choice(x, size = n_samples, p = prob)
    plt.hist(nums, bins = len(x))
    plt.show()
    return nums, prob
```

```python
def uniform(min_, max_, n_samples):
    x = np.arange(min_, max_)
    mean = (min_ + max_)/2
    xU, xL = x + 0.5, x - 0.5 
    prob = ss.uniform.cdf(xU, loc = mean, scale = 100) - ss.uniform.cdf(xL, loc = mean, scale = 100)
    prob = prob / prob.sum() #normalize the probabilities so their sum is 1
    nums = np.random.choice(x, size = n_samples, p = prob)
    plt.hist(nums, bins = len(x))
    plt.show()
    return nums
```

```python
def normal_distribution(mean, sigma, x):
        exponent = -float(((x-mean)**2))/(2*(sigma**2))
        return (1/(np.sqrt(2*3.14*(sigma**2)))) * np.exp(exponent)
```

```python
def plot_samples(xmin, xmax, samples, mean, sigma):
    plt.hist(samples, bins = 300)
    plt.show()
```

```python
def gaussian2(mean, sigma, N):
    accepted = 0 # the number of accepted samples
    samples = np.zeros(N)
    count = 0 # the total count of proposals

    while (accepted < N):
        x = np.random.normal(mean-sigma, mean+sigma)
        y = np.random.normal(0,1)/100
        if y < normal_distribution(mean, sigma, x):
            samples[accepted] = x
            accepted += 1
            
        count +=1
    print("Count", count, "Accepted", accepted)
    
    plot_samples(mean-sigma, mean+sigma, samples, mean, sigma)
    return samples
```

```python
gaussian2(0,100,10000)
```

```python
def normal_distribution_2d(mean, sigma, x):
        exponent = -float(((x-mean)**2))/(2*(sigma**2))
        exponent = float(-((x[0]-mean[0])**2)/(2*sigma[0]**2)-((x[1]-mean[1])**2)/(2*sigma[1]**2))
        return (1/(np.sqrt(2*3.14*(sigma**2)))) * np.exp(exponent)
```
