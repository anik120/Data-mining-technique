```python
import nltk
import enchant
import re
d = enchant.Dict("en_US")
```

```python
from elasticsearch import Elasticsearch
es = Elasticsearch(['http://localhost:9200/'])
doc = {
        'size' : 301,
        'query': {
            'match_all' : {}
       }
   }
res = es.search(index='duc', doc_type='document', body=doc,scroll='1m')
```

```python
kl_rouge = {}
for item in res['hits']['hits']:
    id_ = item['_source']['doc_id']
    gold_summary = item['_source']['gold_summary']
    gold_words = nltk.tokenize.word_tokenize(gold_summary)
    for word in gold_words:
        if not d.check(word):
                gold_words.remove(word)
    kl_summary = item['_source']['kl_summary']
    kl_words = nltk.tokenize.word_tokenize(kl_summary)
    for word in kl_words:
        if not d.check(word):
                kl_words.remove(word)
    no_overlapping_words = 0
    for word in kl_words:
        if word in gold_words:
            no_overlapping_words += 1
    recall = float(no_overlapping_words / len(gold_words))
    precision = float(no_overlapping_words / len(kl_words))
    value = "Recall: " + str(recall) + " Precision: " + str(precision)
    kl_rouge[id_] = value
```

<B> KL Summaries evaluation score </B>

```python
for id_ in kl_rouge:
    print(re.sub('\n','', id_) + ": " + kl_rouge[id_])
    print()
```
