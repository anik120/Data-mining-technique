```python
import scipy.stats as ss
import numpy as np
import matplotlib.pyplot as plt
import sys,os
import random
from collections import Counter
%matplotlib inline
```

```python
def normal_distribution_2d(mean, sigma, x):
        exponent = float(-((x[0]-mean[0])**2)/(2*sigma[0]**2)-((x[1]-mean[1])**2)/(2*sigma[1]**2))
        return (1/2*3.14*sigma[0]*sigma[1] ) * np.exp(exponent)
```

```python
def gaussian2d(mean, sigma, N):
    accepted = 0 # the number of accepted samples
    samples = []
    y_plot = []
    count = 0 # the total count of proposals

    while (accepted < N):
        x1 = np.random.normal(mean[0]-sigma[0], mean[0]+sigma[0])
        x2 = np.random.normal(mean[1]-sigma[1], mean[1]+sigma[1])
        x=[x1,x2]
        y = np.random.normal(0,1)/100
        if y < normal_distribution_2d(mean, sigma, x):
            samples.append(x[0])
            samples.append(x[1])
            y_plot.append(y)
            y_plot.append(y)
            accepted += 1
            
        count +=1
    print("Count", count, "Accepted", accepted)
    plt.scatter(samples, y_plot)
    plt.show()
```

```python
gaussian2d([0,3],[100,100],1000)
```
