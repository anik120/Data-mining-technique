```python
file = open("/Users/droidx/Desktop/Data mining tech/Social mining/ml-100k/u1.base", 'r')
for line in file:
    parts = line.split()
    if parts[0] in user_movies_dict:
        user_movies_dict[parts[0]][parts[1]] = parts[2]
    else:
        user_movies_dict[parts[0]] = {}
        user_movies_dict[parts[0]][parts[1]] = parts[2]
    print(line)
```

```python
def percent_of_matches(user1, user2, user_movies_dict):
    matches = 0
    for movie in user_movies_dict[user1]:
        if movie in user_movies_dict[user2]:
            if user_movies_dict[user1][movie] == user_movies_dict[user2][movie]:
                matches += 1
    return matches / len(user_movies_dict[user1])
```

```python
user_similarity_matrix = np.zeros((944,944))
for user1 in range(1,944):
    for user2 in range(1, 944):
        user_similarity_matrix[user1][user2] = percent_of_matches(str(user1), str(user2), user_movies_dict)
```

```python
user_similarity_dict = {}
for user1 in range(1, 944):
    user_similarity_dict[str(user1)] = []
    for user2 in range(1, 944):
        if user1 != user2:
            match_percent = percent_of_matches(str(user1), str(user2), user_movies_dict)
            if match_percent >= 0.10:
                user_similarity_dict[str(user1)].append(str(user2))
```

```python
test_user_movies_dict = {}
file = open("/Users/droidx/Desktop/Data mining tech/Social mining/ml-100k/u1.test", 'r')
for line in file:
    parts = line.split()
    if parts[0] in test_user_movies_dict:
        test_user_movies_dict[parts[0]][parts[1]] = parts[2]
    else:
        test_user_movies_dict[parts[0]] = {}
        test_user_movies_dict[parts[0]][parts[1]] = parts[2]
```

```python
predicted_user_movie_ratings = {}
for user in test_user_movies_dict:
    predicted_user_movie_ratings[user] = {}
    for movie in test_user_movies_dict[user]:
        rating = 0
        count = 0
        for similar_user in user_similarity_dict[user]:
            if movie in user_movies_dict[similar_user]:
                count += 1
                rating += int(user_movies_dict[similar_user][movie])
        if count != 0:
            predicted_user_movie_ratings[user][movie] = str(rating/count)
        else:
            predicted_user_movie_ratings[user][movie] = str(0)
```

```python
total = 0
correct = 0
for user in test_user_movies_dict:
    for movie in test_user_movies_dict[user]:
        if int(test_user_movies_dict[user][movie]) == math.floor(float(predicted_user_movie_ratings[user][movie])) or test_user_movies_dict[user][movie] == math.ceil(float(predicted_user_movie_ratings[user][movie])):
            correct += 1
        total += 1
print("Accuracy of prediction: " + str((correct/total) * 100))
```
