```python
eps = 3.5
```

```python
neighbors_dict = {}
def create_neighbors():
    for i in range(dataframe.shape[0]):
        neighbors_dict[i] = []
        for j in range(dataframe.shape[0]):
            if(distances[i][j] <= eps):
                neighbors_dict[i].append(j)
        print(len(neighbors_dict[i]))
```
